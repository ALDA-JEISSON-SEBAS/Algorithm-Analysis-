```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../experiments/benchmark_results.csv")

```

```python
avg_df = df.groupby(["Algorithm", "Input Size"])["Time (seconds)"].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(
    data=avg_df,
    x="Input Size",
    y="Time (seconds)",
    hue="Algorithm",
    marker="o"
)

plt.title("Average Execution Time vs Input Size")
plt.xlabel("Input Size (n)")
plt.ylabel("Average Time (seconds)")
plt.grid(True)
plt.show()
```
